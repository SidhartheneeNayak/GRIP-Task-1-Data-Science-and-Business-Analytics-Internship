## Task 1 - Using Linear Regression of two variables to predict student's score based on the number of hours he/she studies per day 

Here I have imported all the required packages for performing linear regression

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
```

Now the .csv file is read into the df dataframe and in x the feature variable is stored and in y the label is stored

```python
df = pd.read_csv('https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv')
print(df.head())
x = np.array(df['Hours'])  # Storing the 'Hours' feature in x as a numpy array
y = np.array(df['Scores']) # Storing the 'Scores' feature in y as a numpy array
```

Now I have plotted a graph between x and y using matplotlib.pyplot to find out a relationship between features and labels

```python
plt.plot(x,y)
plt.xlabel('Hours of study per day')
plt.ylabel('Scores obtained')
plt.show()
```

As we can see that the above line plot does not show a clear relationship between x and y. So below I have plotted a scatter plot between x and y

```python
plt.scatter(x,y)
plt.xlabel('Hours of study per day')
plt.ylabel('Scores of student')
plt.title('Relation between Scores and hours of study')
plt.show()
```

Now from the above scatter plot it is clear that x and y are positively correlated

Below I have splitted the features and label set into training and test set in order to evaluate my model's performance. Then I have reshaped the dataframes into 2d arrays because in the succeeding steps, the function require 2d arrays in order to give a correct output.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=2)#I have used 20% of the data as test set
X_train = X_train.reshape(-1,1)
y_train = y_train.reshape(-1,1)
X_test = X_test.reshape(-1,1)
y_test = y_test.reshape(-1,1)
```

Now I have built the LinearRegression() model and trained it using the training data. Then I have passed the test set to the model for prediction

```python
model = LinearRegression(fit_intercept=True, normalize=True)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
y_pred = y_pred.reshape(-1,1)
```

```python
print('The value of the coefficient is: ',model.coef_)
```

Below, I have plotted the best fit line along with other data points and from the figure we can say that our model has performed quite well.

```python
plt.scatter(X_train,y_train)
plt.plot(X_test, y_pred, color='red')
plt.xlabel('Hours of study per day')
plt.ylabel('Scores of student')
plt.title('Relation between Scores and hours of study')
plt.show()
```

Now I have calculated the mean absolute error between the actual test labels and the predicted labels. Since the MAE is around 4.88 we can say that our model's performance is quite well.

```python
print("The predicted values are: ",y_pred)
print("The actual values are: ",y_test)
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print('Mean absolute error = ',mae)
```
